```python
import numpy as np
from scipy.integrate import simps
import matplotlib.pyplot as plt
```

```python
# Sample voltage data
V = np.linspace(0, 0.5, 100)

# Generate two different exponential IV curves
I1 = (5.728e-6)*np.exp(16.92*V) - (6.113e-4)
I2 = (1.305e-7)*np.exp(21.4*V) - (4.733e-4)
```

```python
# Calculate RMSE
rmse = np.sqrt(np.mean((I1 - I2) ** 2))
print(f"RMSE: {rmse}")

# Calculate MAE
mae = np.mean(np.abs(I1 - I2))
print(f"MAE: {mae}")

# Calculate NRMSE
nrmse = rmse / (np.max(I1) - np.min(I1))
print(f"NRMSE: {nrmse}")

# Calculate Integral of Absolute Differences
integral_diff = simps(np.abs(I1 - I2), V)
print(f"Integral of Absolute Differences: {integral_diff}")

# Calculate AUC Difference
auc_diff = np.abs(simps(I1, V) - simps(I2, V))
print(f"AUC Difference: {auc_diff}")

# Calculate Standard Deviation
std_diff = I2 - I1
std = np.std(std_diff)
print(f"STD Differene: {std}")

# Plot the IV curves
plt.plot(V, I1, label='New')
plt.plot(V, I2, label='Old')
plt.xlabel('Voltage (V)')
plt.ylabel('Current (I)')
plt.legend()
plt.show()
```
